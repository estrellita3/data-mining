Dataset sonar.all-data bajado de:
https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks)

```r
df_sonar <- read.csv('./data/sonar.all-data', header=FALSE)
```

```r
head(df_sonar)
```

```r
colnames(df_sonar)[61]<-'sonar'
```

```r
head(df_sonar)
```

```r
table(df_sonar$sonar)
```

```r
df_sonar$sonar <- factor(df_sonar$sonar)
```

```r
summary(df_sonar)
```

```r
idx <- sample(1:nrow(df_sonar), nrow(df_sonar)*0.7)
df_train<-df_sonar[idx,]
df_test<-df_sonar[-idx,]
```

```r
table(df_train$sonar)
table(df_test$sonar)
```

```r
model <- glm(df_train, formula=sonar~., family = binomial)
summary(model)
```

```r
df_train$pred <- predict(model, df_train)
df_train$pred_sonar <- factor(ifelse(df_train$pred<0,'M','R'))
```

```r
head(df_train[,c('pred','pred_sonar','sonar')])
```

```r
library(ggplot2)
ggplot(df_train, aes(x=pred, color=sonar))+geom_density(bw=5)
```

```r
table(pred=df_train$pred_sonar, real=df_train$sonar)
```

```r
# Ejecutamos en test
df_test$pred <- predict(model, df_test)
df_test$pred_sonar <- factor(ifelse(df_test$pred<0,'M','R'))
table(pred=df_test$pred_sonar, real=df_test$sonar)
```

Para reducir el overfitting vamos a probar a utilizar PCA

```r
model_pca <- prcomp(df_train[,1:60])
```

```r
df_train_pca <- predict(model_pca, df_train)
#df_train_pca <- model_pca$x
```

```r
plot(cumsum(model_pca$sdev^2/sum(model_pca$sdev^2)))
grid()
```

```r
# Nos quedamos con los 20 primeros componentes principales 
df_train_pca <- data.frame(df_train_pca[,1:32])
df_train_pca$sonar <- df_train$sonar
```

```r
model_glm2 <- glm(df_train_pca, formula=sonar~., family = binomial)
summary(model_glm2)
```

```r
df_train_pca$pred <- predict(model_glm2, df_train_pca)
df_train_pca$pred_sonar <- as.factor(ifelse(df_train_pca$pred<0, 'M','R'))
head(df_train_pca[,c('pred','pred_sonar','sonar')])
```

```r
ggplot(df_train_pca, aes(x=pred, color=sonar))+geom_density(bw=0.1)
```

```r
table(pred=df_train_pca$pred_sonar, real=df_train_pca$sonar)
```

```r
df_test_pca <- data.frame(predict(model_pca, df_test[,1:60]) )
df_test_pca$sonar <- df_test$sonar
```

```r
df_test_pca$pred <- predict(model_glm2, df_test_pca)
df_test_pca$pred_sonar <- as.factor(ifelse(df_test_pca$pred<0, 'M','R'))
table(pred=df_test_pca$pred_sonar, real=df_test_pca$sonar)
```

```r
### Ejemplo regularización
```

```r
cvfit<-glmnetUtils::cv.glmnet(sonar~.,
                              family = "binomial",
                              data=df_train[,1:61],nfolds=10,alpha=0.5)
plot(cvfit)
```

```r
df_test$pred<-predict(cvfit,newdata=df_test,s=cvfit$lambda.min)
df_test$pred_sonar <- as.factor(ifelse(df_test$pred<0, 'M','R'))
table(pred=df_test$pred_sonar, real=df_test$sonar)
```
